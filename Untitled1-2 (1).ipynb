```python


# selections des donnees estimees pertinentes
columns_to_keep = ['Socioeconomic Score', 'Study Hours' , 'Attendance (%)', 'Grades']
df = df[columns_to_keep]

print("\nColonnes conservées :")
print(df.columns)
```

# Data cleaning ou nettoyage des données

```python
# Étape 1 : Vérifier et remplacer les valeurs manquantes
print("Valeurs manquantes avant nettoyage :")
print(df.isnull().sum())

# Supprimer les doublons si nécessaire
df = df.drop_duplicates()

# Remplacer les valeurs manquantes par la moyenne de chaque colonne
df = df.fillna(df.mean())

print("\nValeurs manquantes après nettoyage :")
print(df.isnull().sum())

# Étape 2 : Supprimer les valeurs aberrantes manuellement
# Suppression des valeurs aberrantes logiques pour chaque colonne
df = df[(df['Socioeconomic Score'] >= 0) & (df['Socioeconomic Score'] <= 1)]
df = df[(df['Study Hours'] >= 0) & (df['Study Hours'] <= 24)]  # Max 24 heures par jour
df = df[(df['Attendance (%)'] >= 0) & (df['Attendance (%)'] <= 100)]  # Entre 0 et 100%
df = df[(df['Grades'] >= 0) & (df['Grades'] <= 100)]  # Notes entre 0 et 100

# Étape 3 : Vérification finale
print("\nTaille du dataset après nettoyage :", df.shape)
print("\nAperçu des données nettoyées :")
print(df.head())
```

# Data Transformation

```python
#renommination des colonnes pour faciliter leurs compréhension 
df.rename(columns={'Socioeconomic Score': 'statut socioeconomique'}, inplace=True)
df.rename(columns={'Study Hours': 'nmheure detude par jour'}, inplace=True)
df.rename(columns={'Attendance (%)': 'presence en classe(%)'}, inplace=True)
df.rename(columns={'Grades': 'notes obtenues/100'}, inplace=True)
df.columns
```

# Data mining 

```python
# Séparer les caractéristiques (X) et la cible (y)
X = df.drop(columns=["notes obtenues/100"])  # Caractéristiques
y = df["notes obtenues/100"]  # Cible
```

```python
from sklearn.model_selection import train_test_split

# Division en ensembles d'entraînement (train) et de test (test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Vérifier les dimensions des ensembles
print("Entraînement :", X_train.shape, y_train.shape)
print("Test :", X_test.shape, y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

# Initialiser le modèle
model = LinearRegression()

# Entraîner le modèle sur l'ensemble d'entraînement
model.fit(X_train, y_train)
```

# pattern evaluation 

## algo reg.linéaire

```python
# Faire des prédictions sur l'ensemble de test
y_pred = model.predict(X_test)
```

```python
# Afficher les premières prédictions
print("Premières prédictions :", y_pred[:10])
```

```python
# Afficher les premières prédictions
print("Premières prédictions :", y_pred[:10])
```

```python
# Utiliser la méthode score pour obtenir le R²
r2_score = model.score(X_test, y_test)
r2_score
```

```python
# Séparer les caractéristiques (X) en excluant "notes obtenues/100" et "nmheure de sommeil par jour"
a = df.drop(columns=["notes obtenues/100"])

# Séparer la cible (y), qui est "notes obtenues/100"
b = df["notes obtenues/100"]
```

```python
a_train, a_test, b_train, b_test = train_test_split(a, b, test_size=0.2, random_state=42)
```

```python
from sklearn.model_selection import train_test_split

# Diviser les données en ensembles d'entraînement (80%) et de test (20%)
a_train, a_test, b_train, b_test = train_test_split(a, b, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor

# Initialiser le modèle Random Forest
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)

# Entraîner le modèle avec les données d'entraînement
rf_model.fit(a_train, b_train)
```

```python
# Faire des prédictions avec l'ensemble de test
b_pred = rf_model.predict(a_test)
```

```python
from sklearn.metrics import  r2_score

# Calculer le R²
r2 = r2_score(b_test, b_pred)

print(f'R²: {r2}')# Faire des prédictions avec l'ensemble de test
b_pred = rf_model.predict(a_test)
```

# knowlege Presentation

# visualisation de la prediction via la regression lineaire

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Créer une figure
plt.figure(figsize=(8, 6))

# Créer le graphique de dispersion
sns.scatterplot(x=y_test, y=y_pred, alpha=0.6)

# Ajouter une ligne de référence (ligne idéale)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--')

# Ajouter des titres et des labels
plt.title("Vraies valeurs vs Prédictions", fontsize=14)
plt.xlabel("Vraies valeurs (notes obtenues)", fontsize=12)
plt.ylabel("Valeurs prédites (notes obtenues)", fontsize=12)

plt.show()
```

```python
# Calculer les erreurs
errors = y_test - y_pred

# Créer une figure
plt.figure(figsize=(8, 6))

# Créer un histogramme avec Seaborn pour afficher la distribution des erreurs
sns.histplot(errors, kde=True, color='blue', bins=20)

# Ajouter un titre et des labels
plt.title("Distribution des erreurs de prédiction", fontsize=14)
plt.xlabel("Erreur (Vrai - Prédit)", fontsize=12)
plt.ylabel("Fréquence", fontsize=12)

plt.show()
```

```python
# Créer une figure
plt.figure(figsize=(8, 6))

# Créer une heatmap des erreurs absolues
sns.heatmap(abs(errors).values.reshape(-1, 1), cmap='viridis', annot=False, cbar=True)

# Ajouter un titre
plt.title("Heatmap des erreurs absolues", fontsize=14)

plt.show()
```
